## Ejercicios de Limpieza

1. Después de la unión de datos, tenemos dos columnas de "country". Elimina una de ellas.


```python
# Eliminamos una de las columnas Country ya que ambas tienen los mismos valores pero distinto nombre de columna: 
df_merge2.drop(columns=["country"], inplace=True)
df_merge2.head(2)
```


2. Los nombres de las columnas no son homogeneos. Cambia los nombres de las columnas de tal forma que:

    - No tengan espacios.

    - Estén en minúscula.

    - No tengan paréntesis, es decir, quitar "(%)", "(Km2)".

    - Algunas columnas tiene "\n". Eliminalos de los nombres de las columnas.

    - Algunas columnas tienen ":". Eliminalos de los nombres de las columnas.


```python
# Utilizando list comprehensions reemplazamos los espaciosy términos/símbolos que no queremos por NADA. Colocamos todos los nombres de columnas en minúscula también. 
df_colu = [columna.lower().replace(" ", "").replace("(%)", "").replace("(km2)", "").replace("(p/km2)", "").replace("\n", "")
           for columna in df_merge2.columns]

# Asignamos los nuevos nombres de las columnas al DataFrame
df_merge2.columns = df_colu

# Comprobamos que los cambios se han realizado correctamente 
df_merge2.head()
```


3. La columnas coordinates tiene la latitud y la longitud en una sola columna. Crea dos columnas nuevas, una con la longitud y otra con la latitud. Una vez hecho, elimina la columna de coordinates.


```python
# Podemos realizarlo de dos formas: 
# Una forma es con extract que nos extrae dichos valores de las filas correspondientes que cumplan el patrón y nos lo separa en columnas: 
# df_split = df_merge2["coordinates"].str.extract(r"'(.*?)',\s*'(.*?)'")
# df_split.head()
```

```python
# Otra forma más sencilla es con split: 
# Dividir la columna "coordinates" en dos columnas usando la coma como delimitador
df_split = df_merge2["coordinates"].str.split(",", expand=True)
df_split[0] = df_split[0].str.replace("('", "").str.replace("'", "")
df_split[1] = df_split[1].str.replace("')", "").str.replace("'", "")
```

```python
#Visualizamos el data frame 
df_split.head()
```

```python
# Eliminamos la columna coordinates y luego concatenamos ambos data frames: 
df_merge2.drop(columns= ['coordinates'], inplace=True)
df_concat = pd.concat([df_merge2, df_split], axis=1)

# Renombramos los nombres de las columnas y comprobamos: 
df_completo = df_concat.rename(columns={0:'latitud', 1:'longitud'})
df_completo.head()
```


4. Las columnas unemployment_rate, total_tax_rate, tax_revenue, population_labor_force_participation, out_of_pocket_health_expenditure, gross_tertiary_education_enrollment, gross_primary_education_enrollment, forested_area, cpi_change, agricultural_land tienen "%". Elimina los "%" de los valores de las columnas.


```python
#Reemplazamos de los valores que contenan al principio de dicho valor el símbolo concreto por nada: 
df_completo = df_completo.replace('%','', regex= True)
df_completo.head()
```


5. Haz lo mismo para las columnas de gasoline_price, gdp, minimum_wage, pero eliminando "$".


```python
#Reemplazamos de los valores que contenan al principio de dicho valor el símbolo concreto por nada: 
df_completo = df_completo.replace('\$','', regex= True)
df_completo.head()
```

## Ejercicios de Filtrado

Encuentra todos los países cuya mortalidad infantil esté entre 40 y 50 personas por kilómetro cuadrado.


```python
mortalidad = df_completo[df_completo['infantmortality'].between(40,50)]
print('Los países que cumplen la condición:')
print(mortalidad['country'] )
len(mortalidad['country'])
```


Encuentra los países cuyas tasas de natalidad son mayores o iguales a 20 y su esperanza de vida es mayor de 75 años.


```python
natalidad = df_completo[(df_completo['birthrate'] >= 20) & (df_completo['lifeexpectancy'] > 75)]
print('Los países que cumplen la condición:')
print(natalidad['country'] )
len(natalidad['country'])
```


Encuentra las ciudades cuyos paises contienen la palabra "la" en su nombre.


```python
nombre_la = df_completo[df_completo['country'].str.contains('la', case=False, na=False)]
print('Los países que cumplen la condición:')
print(nombre_la['capital/majorcity'] )
print('Cantidad de ciudades que cumplen la condición:', len(nombre_la))
```


Encuentra los países cuyos medicos por cada 1000 habitantes (physicians_per_thousand) sea mayores de 5.


```python
medicos = df_completo[df_completo['physiciansperthousand'] > 5]
print('Los países que cumplen la condición:')
print(medicos['country'])
print('Cantidad de países que cumplen la condición:', len(medicos['country']))
```


Encuentra los países cuyatasa de fertilidad sea mayor a 6.


Encuentra los países cuya moneda es el euro (EUR) y tienen una tasa de natalidad superior al promedio.


```python
moneda = df_completo[(df_completo['currency-code'] == 'EUR') & (df_completo['birthrate'] > (df_completo['birthrate'].mean()))]
print('Los países que cumplen la condición:')
print(moneda['country'])
print('Cantidad de países que cumplen la condición:', len(moneda['country']))
```
