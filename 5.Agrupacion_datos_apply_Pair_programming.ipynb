```python
# Creamos una lista de columnas con las que vamos a trabajar y aplicar la función: 
lista_colum = ['landarea', 'armedforcessize', 'co2-emissions', 'gdp', 'population', 'urban_population', 'latitud', 'longitud', 'density', 'cpi']
def cambio_datos(col):

    try: 
        # Pedimos que nos devuelva en tipo FLOAT los valores de las columnas reemplazando ',' por NADA, ya que hay numeros grandes que utilizan la coma como punto indicando millones y cifras altas:
        return float(col.replace(',','').strip())
    
    except:
        # En el except nos devuelve las columnas, no  np.nan
        return np.nan

#Iteramos por las columnas de nuestra lista de columnas:
for col in lista_colum:    
    
    df_world[col] = df_world[col].apply(cambio_datos)
    
df_world.head()
```

```python
# Creamos la función continente que asigna un país (df_world['country']) a su continente utilizando un diccionario:
def continente(diccionario, pais):
    try:
        # Intentamos acceder al continente correspondiente al país en el diccionario
        return diccionario[pais]
    except KeyError:
        # Si el país no se encuentra en el diccionario, devuelve NaN
        return np.nan

# Utilizamos el método apply() en la columna 'country' del DataFrame df_world
# Aplica una función lambda a cada valor de esa columna para asignar el continente correspondiente:
df_world['continent'] = df_world['country'].apply(lambda pais: continente(paises_continente, pais))
```


¿Cuál es la tasa de impuestos total máxima por densidad de población por continente? Para solucionar este ejercicio puedes:



Primero, necesitas calcular la tasa de impuestos total para cada continente. La tasa de impuestos total se encuentra en la columna "total_tax_rate".



Luego, calcula la densidad de población para cada continente. La densidad de población se encuentra en la columna "density".



Ahora, divide la tasa de impuestos total entre la densidad de población para obtener la tasa de impuestos total por persona por kilómetro cuadrado.


```python
# Agrupamos por continente
continente_group = df_world.groupby('continent')
# Calculamos maximo impuestos total y promedio de densidad. Dividimos: 
impuestos_km2 = continente_group['totaltaxrate'].max()/continente_group['density'].mean()
# Lo sacamos como dataframe 
df_impuestos = impuestos_km2.reset_index()
# Cambiamos el nombre de la columna  0 por una que consideramos que mejora la lectura del dataframe 
df_impuestos.rename(columns={0: 'impuesto_km2'}, inplace=True)

df_impuestos.head()
```


Encuentra el país que tiene el valor más alto en esta nueva columna calculada.


```python
# Agrupamos por país
pais_group = df_world.groupby('country')
# Calculamos maximo impuestos total y promedio de densidad. Dividimos: 
impuestos_pais = pais_group['totaltaxrate'].max()/pais_group['density'].mean()
# Lo sacamos como dataframe 
df_impuestos_pais = impuestos_pais.reset_index()
# Cambiamos el nombre de la columna  0 por una que consideramos que mejora la lectura del dataframe
df_impuestos_pais.rename(columns={0: 'impuesto_km2'}, inplace=True)

df_impuestos_pais.head()
```


¿Cuál es la tasa de maternidad promedio por tasa de impuestos total? Para resolver el ejercicio deberás:



Primero, calcula la tasa de maternidad promedio para cada continente. Puedes usar la columna "maternal_mortality_ratio" que representa la tasa de mortalidad materna.



Calcula la tasa de impuestos total promedio para cada continente. Puedes usar la columna "total_tax_rate" que representa la tasa de impuestos total en cada continente.


```python
# Agrupamos por continente y calculamos el promedio de maternal mortality ratio:
continentes = df_world.groupby('continent')
maternidad = continentes['maternalmortalityratio'].mean()

#Calculamos promedio de impuestos total:
impuestos = continentes['totaltaxrate'].mean()
```


Luego, divide la tasa de maternidad promedio por la tasa de impuestos total promedio para obtener la relación entre ambas tasas.


```python
relacion_tasas = maternidad/impuestos
df_relacion = relacion_tasas.reset_index()
df_relacion.rename(columns={0:'relacion_tasa'},inplace=True)
df_relacion.head()
```
